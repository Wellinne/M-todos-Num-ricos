# Metodologia

**O modelo SIRD**

Foi em 1927 que W. O. Kermack e A. G. McKendrick publica-ram o artigo intitulado“A Contribution to the Mathematical Theory of Epidemics”, no qual introduziram o modelo epi-demiológico conhecido como SIR. Esse modelo descreve a disseminação de doenças infecciosas em uma população dividida em subgrupos de indivíduos susceptíveis, infectados e recuperados, com cada um desses grupos tendo sua dinâmica descrita por uma Equação Diferencial Ordinária(EDO). 

O modelo SIRD é uma extensão do modelo SIR com a adição de duas premissas que são a taxa de recuperação com imunidade e a taxa morte.

*   Taxa de mortalidade (mi)   
*   Taxa de infecção    (beta)


# Resultados e Discussão



  Seguindo nossa metodologia de utilizar o modelo SIRD, e assim feito no código que será visto abaixo, foi necessário definir cada valor de entrada, que são representadas pelas variáveis do código. Agora, vamos fazer um detalhamento de quais parâmetros serão utilizados e qual é o motivo de sua utilização, cada variável citada a seguir está definida no código.

* *N* - Tem como valor de entrada em todas as plotagens o número de habitantes que residem em Pernambuco;

* *I0* - Representa o número de infectados do dia 1 da nossa pesquisa;

* *R0* - Representa o total de recuperados no dia 1;

* *D0* - É a quantidade de pessoas mortas no começo do experimento;

* *S0* - São as pessoas suscetíveis a adquirir a doença;

* *simulation_days* - Onde é definido a quantidade de dias que queremos prever;

* *beta* - É a nossa taxa de infecção, que é calculada pela razão entre  o total de recuperados inicialmente e a população de pernambuco;  

* *gama* - É a taxa de recuperação, ela é calculada através da razão entre o número de recuperados e a população de pernambuco;

* *mi* - Representa a taxa de mortalidade, dada pela razão entre o número de mortos inicialmente e a população inicial.



  Sabendo disso, como o intuito da discussão em questão é mostrar a importância da vacina, decidimos fazer uma comparação de como seria a evolução do vírus nas pessoas que tomaram a vacina e nas pessoas que não tomaram. Para que isso fosse possível, começamos a previsão utilizando os dados do dia 31 de dezembro de 2021 (não iniciamos em 1 de janeiro de 2022 como foi pedido pois os dados estariam subnotificados visto que é um dia atípico de feriado e as unidades que reportam os casos e óbitos não estariam funcionando plenamente, e a subnotificação iria interferir na nossa projeção), pois a partir desse período a vacinação já estava consolidada em Pernambuco, então modelamos até o dia 31 de julho de 2022 (que foi a projeção pedida na devolutiva do nosso P3) com base nos dados iniciais (31 de dezembro). 

 Faremos a projeção para o mesmo intervalo de dias que a previsão anterior de 31/12/2021-31/07/2022 (212 dias) tendo como dados iniciais uma data que não havia vacina, também projeção da variante delta e gama, sempre no mesmo intervalo de 212 dias, visto que uma das coisas que repercutiu bastante no contexto pandêmico foi a chegada das novas variantes, e a cada nova chegada sempre se iniciava uma nova onda e refletiremos qual foi ou teria sido a importância da vacina.





# Código

```python
#Importando bibiotecas
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
%matplotlib inline 
```

# Previsão de 31/12/2021 até 31/07/2022 (ômicron)

```python
#def Variavéis

#População inicial (população de pernambuco)
N =  9616621 
#total de infectados (quantidade de confirmados)
I0 = 645702
#total de recuperados 
R0 = 583828
#total de mortos
D0 = 20447
#população Suscetivél
S0 = (N - I0 - R0- D0)
 
#Taxa de infecção
beta = I0 / N
#taxa de recuperação
gama = R0 / N
#taxa de mortalidade
mi = D0 / N 

#Numero de dias da simulação
simulation_days = 212 

t = np.linspace(0, 220, 220)

#Equações diferenciais do modelo SIRD, que recebe um vetor com S, I, R e D como entrada e retorna suas derivadas

def deriv(y, t, N, beta, gama, mi):
    S, I, R, D = y
    dSdt = - (beta * I * S) / N
    dIdt = ((beta * S * I) / N) - (gama*I) - (mi * I) 
    dRdt = gama * I
    dDdt = mi * I
    return dSdt, dIdt, dRdt, dDdt 

#Resolução do sistema de EDO's

y0 = S0, I0, R0, D0
ret = odeint(deriv, y0, t, args=(N, beta, gama, mi))
S, I, R, D = ret.T

#Exibindo Resultados

fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, S/N, 'b', alpha=0.5, lw=2, label='Susceptible')
ax.plot(t, I/N, 'y', alpha=0.5, lw=2, label='Infected')
ax.plot(t, R/N, 'g', alpha=0.5, lw=2, label='Recovered with immunity')
ax.plot(t, D/N, 'r', alpha=0.5, lw=2, label='Deads')
ax.set_xlabel('Time /days')
ax.set_ylabel('Number (1000s)')
ax.set_ylim(0,1)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(b=True, which='major', c='w', lw=2, ls='-')
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.savefig('sird.pdf')
plt.show()
```

> *É possível observar que o número de recuperados aumenta e o de suscetíveis diminui, chegando a um ponto constante, e o de infectados e de mortes cai drasticamente, mesmo com a ômicron, que já existia nesse período com a vacinação em massa da população, que já era uma realidade, em Pernambuco em 31 de dezembro 2021 cerca de 13.557.398 doses da vacina haviam sido aplicadas em solo Pernambucano, assim, a vacina se fez altamente eficaz, salvando inúmeras vidas, mesmo com a liberação do uso obrigatório da máscara pelas autoridades competentes.*
